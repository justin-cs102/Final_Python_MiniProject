```python
# The purpose of the code is to create a Calculator Program.
# This calculator takes inputs to have calculations done on them
# The operations that can be conducted on them. These include multiplication, subtraction, addition, and division.
# The calculator also has the option to take the reciprocal, sine, and cosine of a number
# If prompted, the calculator can display previous calculations

import math
#List to store each calculated result
results = []

# The following are functions that accept two numbers to have calculations done on them. 
# The functions check if the arithmetic operation works with "try" statement
def plus(n1, n2):
    try:
        return n1 + n2
    except:
        print("Overflow error! ")


def minus(n1, n2):
    try:
        return n1 - n2
    except:
        print("Overflow error! ")


def multiply(n1, n2):
    try:
        return n1 * n2
    except:
        print("Overflow error! ")


def divide(n1, n2):
    try:
        return n1 / n2
    except:
        print("Error! Please put valid numbers.")

def reciprocal(n1):
    try:
        return 1/n1
    except:
        print("Error! Please put valid number.")

def sine(n1):
    try:
        return str(math.sin(n1))
    except:
        print("Error! ")

def cosine(n1):
    try:
        return str(math.cos(n1))
    except:
        print("Error! ")

#Try function that tests if the answers have errors
def check(number):
    try:
        float(number)
    except:
        print("Please type in a valid number")
        num = str(input("Retype number. "))
        return check(num)
    else:
        return float(number)

#This function checks if the input is a string
def canConvert(string):
    try:
        int(string)
    except:
        return False
    else:
        return True

#This function prints the results
def showResults():
    print("Results:")
    for i in results:
        print("\n" + i)




print("Welcome to Calculator!")
while True:
    print(
        "\nWhat would you like to do? (Type the number corresponding to action)\n"
        "1) Add\n2) Subtract\n3) Multiply\n"
        "4) Divide\n5) Reciprocal\n6) Take the sine of a number in radians\n7) Take the cosine of a number in radians\n"
        "8) Show Previous Calculations\n9) Exit\n")
    sign = input()

    while True:
        if not canConvert(sign) or int(sign) < 1 or int(sign) > 10:
            sign = input("Please enter a valid choice (1-9).\n\nWhat would you like to do? (Type the number corresponding to action)\n"
        "1) Add\n2) Subtract\n3) Multiply\n"
        "4) Divide\n5) Reciprocal\n6) Take the sine of a number in radians\n7) Take the cosine of a number in radians\n"
        "8) Show Previous Calculations\n9) Exit\n")
        else:
            break
#The calculator will exit or show previous calculations when prompted
    output = 0
    if int(sign) == 9:
        print("Goodbye! ")
        exit(0)

    if int(sign) == 8:
        showResults()
        continue

    if int(sign) == 6:
        num1 = str(input("Type number 1: "))
        num1 = check(num1)
        output = "The sine of " + str(num1) + " is  equal to " + sine(num1)
        print(output)

    if int(sign) == 7:
        num1 = str(input("Type number 1: "))
        x = str(math.cos(num1))
        output = "The cosine of " + str(num1) + " is  equal to " + x
        print(output)




    # if users wishes, the algorithm will add/subtract/multiply/divide/find the reciprocal those numbers
    if int(sign) == 1:
        num1 = str(input("Type number 1: "))
        # asks for inputs, and floats ensure the python reads the input as a number with decimals
        num1 = check(num1)
        num2 = str(input("Type number 2: "))
        num2 = check(num2)
        output = str(num1) + " + " + str(num2) + " = " + str(plus(num1, num2))
        print(output)


    elif int(sign) == 2:
        num1 = str(input("Type number 1: "))
        # asks for inputs, and floats ensure the python reads the input as a number with decimals
        num1 = check(num1)
        num2 = str(input("Type number 2: "))
        num2 = check(num2)
        output = str(num1) + " - " + str(num2) + " = " + str(minus(num1, num2))
        print(output)

    elif int(sign) == 3:
        num1 = str(input("Type number 1: "))
        # asks for inputs, and floats ensure the python reads the input as a number with decimals
        num1 = check(num1)
        num2 = str(input("Type number 2: "))
        num2 = check(num2)
        output = str(num1) + " * " + str(num2) + " = " + str(multiply(num1, num2))
        print(output)

    elif int(sign) == 4:
        num1 = str(input("Type number 1: "))
        # asks for inputs, and floats ensure the python reads the input as a number with decimals
        num1 = check(num1)
        num2 = str(input("Type number 2: "))
        num2 = check(num2)
        output = str(num1) + " / " + str(num2) + " = " + str(divide(num1, num2))
        print(output)

    elif int(sign) == 5:
        num1 = str(input("Type number 1: "))
        # asks for inputs, and floats ensure the python reads the input as a number with decimals
        num1 = check(num1)
        output = "1 /" + str(num1) + " = " + str(reciprocal(num1))
        print(output)
#Appends/adds each result to "Results" list
    results.append(output)
```
